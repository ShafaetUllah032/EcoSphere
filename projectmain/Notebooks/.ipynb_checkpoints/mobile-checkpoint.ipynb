```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
dataset=pd.read_csv('mobile_data/train.csv')
```

```python
datas=dataset
```

```python
correlation = datas.corr()

plt.figure(figsize=(20, 20))
sns.heatmap(correlation,vmin=-1, vmax=1, center=0,
cmap=sns.diverging_palette(50, 500, n=500),square=True,annot=True)
```

```python
dataset.info()
```

```python
dataset
```

```python
y=dataset['price_range']
```

```python
dataset=dataset.drop(["price_range"],axis=1)
dataset=dataset.drop(['blue'],axis=1)
dataset=dataset.drop(['m_dep'],axis=1)
dataset=dataset.drop(['n_cores'],axis=1)
dataset=dataset.drop(['pc'],axis=1)
dataset=dataset.drop(['px_height'],axis=1)
dataset=dataset.drop(['px_width'],axis=1)
dataset=dataset.drop(['sc_h'],axis=1)
dataset=dataset.drop(['sc_w'],axis=1)
dataset=dataset.drop(['three_g'],axis=1)
dataset=dataset.drop(['fc'],axis=1)
dataset=dataset.drop(['clock_speed'],axis=1)

```

```python
dataset.info()
```

```python
X=dataset
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=101)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred.round(), normalize=True)
```

```python
from joblib import dump, load
```

```python
dump(model, './../savedModels/model.joblib')
```
